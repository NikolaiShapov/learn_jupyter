```python
import csv
```

```python
streets = {}
bad_streets = []
```

```python
%%time
with open ('bus_stops.csv' , 'r', encoding = 'cp1251') as f:
    reader = csv.DictReader(f, delimiter =';')
    for row in reader:
        znak = row['PlaceDescription'].find(',')
        if znak != -1 or len(row) != 0:
            if row['PlaceDescription'][:znak] in streets:
                streets[row['PlaceDescription'][:znak]] += 1
            else:
                streets[row['PlaceDescription'][:znak]] = 1
        else:
            bad_streets.append(row['PlaceDescription'])
```
